```python
dataF.fillna(value='None',inplace=True)
sns.heatmap(dataF.isnull(),yticklabels=False,cbar=False)
```

```python
arr = dataF['age'].values

arr_min = []
arr_max = []

for item in arr:
    arpp = item.split('-')
    arr_min.append(arpp[0])
    arr_max.append(arpp[1])
```

```python
dataF['age_min'] = pd.Series(arr_min, index=dataF.index)
```

```python
dataF['age_max'] = pd.Series(arr_max, index=dataF.index)
```

```python
drt= dataF.groupby(['campaign_id','reporting_date'])['reporting_date'].count()
df = pd.DataFrame(drt.unstack(level=1))
df.fillna(value=0,inplace=True)
df = df.transpose()
df['Total Ads'] = df.sum(axis=1)
dt = df['Total Ads']
dt = pd.DataFrame(dt)
dx = dt.to_csv('D:/abc.csv')
dx = pd.read_csv('D:/abc.csv')
repo_Tot_ads = dx

drt= dataF.groupby(['campaign_id','reporting_date'])['spent'].sum()
df = pd.DataFrame(drt.unstack(level=1))
df.fillna(value=0,inplace=True)
df = df.transpose()
df['Total Spent'] = df.sum(axis=1)
dt = df['Total Spent']
dt = pd.DataFrame(dt)
dx = dt.to_csv('D:/abc.csv')
dx = pd.read_csv('D:/abc.csv')
repo_Tot_spent = dx

drt= dataF.groupby(['campaign_id','reporting_date'])['impressions'].mean()
df = pd.DataFrame(drt.unstack(level=1))
df.fillna(value=0,inplace=True)
df = df.transpose()
df['Total Impressions'] = df.sum(axis=1)
dt = df['Total Impressions']
dt = pd.DataFrame(dt)
dx = dt.to_csv('D:/abc.csv')
dx = pd.read_csv('D:/abc.csv')
repo_Tot_impr = dx

drt= dataF.groupby(['campaign_id','reporting_date'])['interest1'].sum()
df = pd.DataFrame(drt.unstack(level=1))
df.fillna(value=0,inplace=True)
df = df.transpose()
df['Total Interests'] = df.sum(axis=1)
dt = df['Total Interests']
dt = pd.DataFrame(dt)
dx = dt.to_csv('D:/abc.csv')
dx = pd.read_csv('D:/abc.csv')
repo_Tot_intr1 = dx


drt= dataF.groupby(['campaign_id','reporting_date'])['clicks'].count()
df = pd.DataFrame(drt.unstack(level=1))
df.fillna(value=0,inplace=True)
df = df.transpose()
df['Total Clicks'] = df.sum(axis=1)
dt = df['Total Clicks']
dt = pd.DataFrame(dt)
dx = dt.to_csv('D:/abc.csv')
dx = pd.read_csv('D:/abc.csv')
repo_Tot_clicks = dx

dataF['age_min'] = dataF['age_min'].convert_objects(convert_numeric=True)
dataF['age_max'] = dataF['age_max'].convert_objects(convert_numeric=True)

drt= dataF.groupby(['campaign_id','reporting_date'])['age_min'].mean()
df = pd.DataFrame(drt.unstack(level=1))
df.fillna(df.mean(),inplace=True)
df = df.transpose()
df['Total MinAge'] = df.mean(axis=1)
dt = df['Total MinAge']
dt = pd.DataFrame(dt)
dx = dt.to_csv('D:/abc.csv')
dx = pd.read_csv('D:/abc.csv')
repo_Tot_MinAge = dx

drt= dataF.groupby(['campaign_id','reporting_date'])['age_max'].mean()
df = pd.DataFrame(drt.unstack(level=1))
df.fillna(df.mean(),inplace=True)
df = df.transpose()
df['Total MaxAge'] = df.mean(axis=1)
dt = df['Total MaxAge']
dt = pd.DataFrame(dt)
dx = dt.to_csv('D:/abc.csv')
dx = pd.read_csv('D:/abc.csv')
repo_Tot_MaxAge = dx
```
