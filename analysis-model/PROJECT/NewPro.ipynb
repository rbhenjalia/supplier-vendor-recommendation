```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from ast import literal_eval
from collections import defaultdict
%matplotlib inline
import random
from sklearn.preprocessing import MinMaxScaler
```

```python
dataFrame = pd.read_csv('D:/BestSellers.csv')
dataFrame = dataFrame[['Product', 'Price', 'Count', 'Buyer_FeedBack','Courier_Rating', 'Product_Rating', 'Packaging_Rating', 'ALL OVER']]
data = pd.read_csv('D:/FinalDataset.csv')
data = data[['Product', 'Brand']]
dataFrame = dataFrame.merge(data,how='left')
dataFrame = dataFrame.drop_duplicates(keep="first")
dataFrame.head()
dataFrame.columns
```

```python
dr = pd.DataFrame(dataFrame.groupby(['Brand','Price'])['Price'].count())
dr = dr.unstack(level=0).T
dr = pd.DataFrame(dr.idxmax(axis=1)).reset_index()
dr = dr.rename(columns={0:'Price'})
dr = dr[['Brand','Price']]
Price = dr

dr = pd.DataFrame(dataFrame.groupby(['Brand','Count'])['Count'].count())
dr = dr.unstack(level=0).T
dr = pd.DataFrame(dr.idxmax(axis=1)).reset_index()
dr = dr.rename(columns={0:'Count'})
dr = dr[['Brand','Count']]
Count = dr

dr = pd.DataFrame(dataFrame.groupby(['Brand','Buyer_FeedBack'])['Buyer_FeedBack'].count())
dr = dr.unstack(level=0).T
dr = pd.DataFrame(dr.idxmax(axis=1)).reset_index()
dr = dr.rename(columns={0:'Buyer_FeedBack'})
dr = dr[['Brand','Buyer_FeedBack']]
Buyer_FeedBack = dr

dr = pd.DataFrame(dataFrame.groupby(['Brand','Courier_Rating'])['Courier_Rating'].count())
dr = dr.unstack(level=0).T
dr = pd.DataFrame(dr.idxmax(axis=1)).reset_index()
dr = dr.rename(columns={0:'Courier_Rating'})
dr = dr[['Brand','Courier_Rating']]
Courier_Rating = dr

dr = pd.DataFrame(dataFrame.groupby(['Brand','Product_Rating'])['Product_Rating'].count())
dr = dr.unstack(level=0).T
dr = pd.DataFrame(dr.idxmax(axis=1)).reset_index()
dr = dr.rename(columns={0:'Product_Rating'})
dr = dr[['Brand','Product_Rating']]
Product_Rating = dr

dr = pd.DataFrame(dataFrame.groupby(['Brand','Packaging_Rating'])['Packaging_Rating'].count())
dr = dr.unstack(level=0).T
dr = pd.DataFrame(dr.idxmax(axis=1)).reset_index()
dr = dr.rename(columns={0:'Packaging_Rating'})
dr = dr[['Brand','Packaging_Rating']]
Packaging_Rating = dr

dr = pd.DataFrame(dataFrame.groupby(['Brand','ALL OVER'])['ALL OVER'].count())
dr = dr.unstack(level=0).T
dr = pd.DataFrame(dr.idxmax(axis=1)).reset_index()
dr = dr.rename(columns={0:'ALL OVER'})
dr = dr[['Brand','ALL OVER']]
ALL_OVER = dr

f1 = pd.merge(Packaging_Rating, ALL_OVER , on=['Brand'])
f2 = pd.merge(Price, Count, on=['Brand'])
f3 = pd.merge(Courier_Rating, Buyer_FeedBack , on=['Brand'])

f12 =  pd.merge(f1,f2 , on=['Brand'])
f34 =  pd.merge(f3,Product_Rating , on=['Brand'])

finSet = pd.merge(f12,f34 , on=['Brand'])
```

```python
finSet.to_csv('D:/BestSellers_Brand_Wise.csv')
```

```python
finSet = finSet[['Brand','Price', 'Count','Courier_Rating', 'Buyer_FeedBack', 'Product_Rating','Packaging_Rating','ALL OVER']]
```

```python
p = pd.DataFrame(finSet.groupby('Price')['Brand'].count())
p = pd.DataFrame(p.unstack(level=0)).reset_index()
p = p.rename(columns={'Price':'Brand',0:'Price'})
p = p[['Brand','Price']]

c = pd.DataFrame(finSet.groupby('Count')['Brand'].count())
c = pd.DataFrame(c.unstack(level=0)).reset_index()
c = c.rename(columns={'Count':'Brand',0:'Count'})
c = c[['Brand','Count']]

cr = pd.DataFrame(finSet.groupby('Courier_Rating')['Brand'].count())
cr = pd.DataFrame(cr.unstack(level=0)).reset_index()
cr = cr.rename(columns={'Courier_Rating':'Brand',0:'Courier_Rating'})
cr = cr[['Brand','Courier_Rating']]

bf = pd.DataFrame(finSet.groupby('Buyer_FeedBack')['Brand'].count())
bf = pd.DataFrame(bf.unstack(level=0)).reset_index()
bf = bf.rename(columns={'Buyer_FeedBack':'Brand',0:'Buyer_FeedBack'})
bf = bf[['Brand','Buyer_FeedBack']]

pr = pd.DataFrame(finSet.groupby('Product_Rating')['Brand'].count())
pr = pd.DataFrame(pr.unstack(level=0)).reset_index()
pr = pr.rename(columns={'Product_Rating':'Brand',0:'Product_Rating'})
pr = pr[['Brand','Product_Rating']]

par = pd.DataFrame(finSet.groupby('Packaging_Rating')['Brand'].count())
par = pd.DataFrame(par.unstack(level=0)).reset_index()
par = par.rename(columns={'Packaging_Rating':'Brand',0:'Packaging_Rating'})
par = par[['Brand','Packaging_Rating']]

ar = pd.DataFrame(finSet.groupby('ALL OVER')['Brand'].count())
ar = pd.DataFrame(ar.unstack(level=0)).reset_index()
ar = ar.rename(columns={'ALL OVER':'Brand',0:'ALL OVER'})
ar = ar[['Brand','ALL OVER']]

a1 = pd.merge(p,c,on='Brand',how='outer')
a2 = pd.merge(cr,bf,on='Brand',how='outer')
a3 = pd.merge(pr,par,on='Brand',how='outer')

a12 = pd.merge(a1,a2,on='Brand',how='outer')

trpp = pd.merge(a12,a3,on='Brand',how='outer')
trpp = pd.merge(trpp,ar,on='Brand',how='outer')
               
trpp.fillna(value=0,inplace=True)
```
